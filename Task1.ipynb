```python
remove_duplicates=data_filled.drop_duplicates()
remove_duplicates
```

```python
{col: remove_duplicates[col].nunique() - remove_duplicates[col].str.strip().str.lower().str.title().nunique()
 for col in remove_duplicates.select_dtypes(include='object').columns}
```

```python
standardized_data = remove_duplicates.apply(lambda col: col.astype(str).str.strip().str.lower().str.title() if col.dtype == 'object' else col)
```

```python
standardized_data
```

```python
print(standardized_data.select_dtypes(include='object').columns)

```

```python
standardized_data['saledate'] = standardized_data['saledate'].str.extract(r'^(.+?\d{2}:\d{2}:\d{2})')[0] 
standardized_data['saledate'] = pd.to_datetime(standardized_data['saledate'], errors='coerce')
standardized_data['saledate'] = standardized_data['saledate'].dt.strftime('%d-%m-%Y')
```

```python
standardized_data
```

```python
renamed_data = (
    standardized_data.columns
    .str.strip()         
    .str.lower()           
    .str.replace(' ', '_')  
    .str.replace(r'[^\w_]', '', regex=True)  
)
```

```python
renamed_data
```

```python
print(standardized_data.dtypes)
```

```python
standardized_data = standardized_data.apply(
    lambda col: pd.to_datetime(col, errors='coerce', dayfirst=False) if col.name == 'saledate' else pd.to_numeric(col, errors='coerce') if col.name in ['odometer', 'mmr', 'sellingprice'] else col.astype('int64') if col.name == 'year' else col.str.strip().str.lower().str.title() if col.dtype == 'object' else col
)
```

```python
print(standardized_data.dtypes)
```

```python
standardized_data.to_csv("cleaned_data.csv", index=True)
print("\n📁 Cleaned dataset saved as 'cleaned_data.csv'")
```
